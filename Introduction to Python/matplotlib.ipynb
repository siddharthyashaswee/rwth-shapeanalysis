```python
import matplotlib.pyplot as plt
import numpy as np
```

# Plots

## Scatter Plots

```python
X = np.random.rand(1000, 2)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

## Bar Plots, Box Plots & Histograms

```python
X = np.random.rand(5)
plt.bar(np.arange(5), X)
plt.show()
```

```python
data_1 = np.random.rand(100)
data_2 = np.random.rand(100) * 2 + 1
data_3 = np.random.rand(100) * 2 - 1
plt.boxplot([data_1, data_2, data_3])
plt.show()
```

```python
data = np.random.rand(1000)
plt.hist(data, bins=10)
plt.show()
```

```python
plt.hist(data, bins=np.linspace(0, 1, 100))
plt.show()
```

```python
plt.hist(data, bins='fd')
plt.show()
```

## Line Plots

```python
x = np.linspace(-np.pi, np.pi, 1000, endpoint=True)
c = np.cos(x)
s = np.sin(x)
```

```python
#plot for cos function
plt.plot(x, c)
plt.show()
```

## Modifying Plots

```python
plt.plot(x, c)
plt.plot(x, s)
plt.show()
```

```python
plt.plot(x, c)
plt.plot(x, s, color='red')
plt.show()
```

```python
plt.plot(x, c)
plt.plot(x, s, color='red')
plt.ylim(-1,1)
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
```

```python
plt.plot(x, c)
plt.plot(x, s, color='red')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.show()
```

```python
plt.plot(x, c)
plt.plot(x, s, color='red')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1])
plt.show()
```

```python
plt.plot(x, c, label='cosine')
plt.plot(x, s, color='red', label="sine")
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1])
plt.legend(loc='upper left', frameon=False)
plt.show()
```

```python
plt.plot(x, c, label='cosine')
plt.plot(x, s, color='red', label="sine")
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1])
plt.legend(loc='upper left', frameon=False)

ax = plt.gca() # get current axis
ax.spines['right'].set_color('yellow')
ax.spines['top'].set_color('orange')
ax.spines['left'].set_color('green')
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
plt.show()
```

## Visualizing Matrices

```python
A = np.random.rand(10,10)
print(A)
plt.matshow(A)
```

```python
# a 2D array with linearly increasing values on the diagonal
a = np.diag(range(15))

plt.matshow(a)
print(a)
plt.show()
```

```python
A = np.random.rand(3,3,3)

plt.matshow(A)
print(A)
plt.show()
```

```python
#20x20 matrix
B = np.random.randn(20, 20)
#set row 5 and col 12 to 0
B[5, :] = 0.
B[:, 12] = 0.
#Plot the sparsity pattern of a 2D array.
#This visualizes the non-zero values of the array.
plt.spy(B, precision=0)
```

```python
#If precision is 0, any non-zero value will be plotted. Otherwise, values of 
#will |Z|>precision be plotted.
plt.spy(B, precision=0.1)
```
